# Iris Dataset
### Objective: Learn how to load, inspect, and visualize a dataset to understand data trends and distributions.

##### Importing Libraries and loading the dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load the dataset
# Seaborn has it built-in, and it loads as a pandas DataFrame.
iris = sns.load_dataset('iris')
```

##### Describing the dataset to get understanding of it

```python
# 2. Getting to know the data
print(iris.head())     # Displays first 5 rows
print(iris.info())     # Provides info regarding datatypes
print(iris.describe()) # Provides min, max, mean etc

# 3. Your 'Pandas-Only' Challenge Questions:
# (Try to answer these before you plot anything)
iris['species'].value_counts()
iris['species'].value_counts()
```
